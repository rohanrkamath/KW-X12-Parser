```python
# test for stitching back claims into EDI format on claims deletion from original EDI file.
# randomly remove 10 rows from the dataframe; this is to simulate claims being held, which in this case is 10, and the rest are set to be released.

import random

rows_to_drop = random.sample(list(df1.index), 10)
df_reduced = df1.drop(rows_to_drop)

print(f"Original: {df1.shape[0]} rows x {df1.shape[1]} cols")
print(f"Reduced:  {df_reduced.shape[0]} rows x {df_reduced.shape[1]} cols")
df_reduced.head()
```

```python
from kw_x12_parser import write_to_edi_x837p

# original EDI text and filename from the DataFrame
original_text = df_reduced["original_837_edi_test"].iloc[0]
original_filename = df_reduced["source_file"].iloc[0]

write_to_edi_x837p(
    df_reduced,
    original_filename,
    original_edi=original_text,
)

# The result is a new EDI file with the same name as the original file, but with the claims removed from the block above.
```
